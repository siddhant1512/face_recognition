```python
import cv2
import numpy as np
```

```python
cap=cv2.VideoCapture(0)
i=0
face_data=[]
path='./data/'
file=input("enter the name of the person")
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
while True:
    ret,frame=cap.read()
    
    if ret==False:
        continue
    face=face_cascade.detectMultiScale(frame,1.3,5)
    face=sorted(face,key=lambda f:f[2]*f[3])
    
    for face in face[-1:]:
        x,y,w,h=face
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        offset=10
        face_section=frame[y-offset:y+h+offset,x-offset:x+w+offset]
        face_section=cv2.resize(face_section,(100,100))
        
        cv2.imshow('section',face_section)
        i+=1
        if(i%10==0):
            
            face_data.append(face_section)
            print(len(face_data))
    cv2.imshow('frame',frame)
    
    
    key= cv2.waitKey(1) & 0xFF
    if key==ord('q'):
        break
print("face_data",face_data)        
face_data=np.asarray(face_data)
face_data=face_data.reshape((face_data.shape[0],-1))
print(face_data.shape)
np.save(path+file+'.npy',face_data)

    
    
cap.release()
cv2.destroyAllWindows()
```
